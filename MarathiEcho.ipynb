```python
import sys
import webbrowser
import pyautogui
import speech_recognition as sr
import os
import datetime
import wikipedia
from gtts import gTTS
import time
import random
import ctypes
import requests
```

```python
# Set console title and clear screen
ctypes.windll.kernel32.SetConsoleTitleW("Virtual Assistant By Coder")
os.system('cls')
```

```python
def speak(text):
    """Speak the given text using Google TTS."""
    tts = gTTS(text=text, lang='mr')
    tts.save("output.mp3")
    os.system("start output.mp3")
    time.sleep(len(text) * 0.1)
    print(text)
```

```python
def wishme():
    """Wish the user based on the time of day."""
    hour = datetime.datetime.now().hour
    if 0 <= hour < 12:
        speak('शुभ प्रभात सर. तुम्ही कसे आहात?')
    elif 12 <= hour < 16:
        speak('शुभ दुपार सर. तुम्ही कसे आहात?')
    else:
        speak('शुभ संध्याकाळ सर. तुम्ही कसे आहात?')
```

```python
def takecommand():
    """Listen for and recognize a voice command from the user."""
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print('Listening...')
        audio = r.listen(source)
        try:
            text = r.recognize_google(audio, language='mr-IN')
            text = text.lower()
            print('You said:', text)
        except sr.UnknownValueError:
            print('Sorry, I could not understand. Please try again.')
            return ""
        except sr.RequestError as e:
            print(f'Sorry, I could not connect to Google services; {e}')
            return ""
        return text
```

```python
def get_weather(city):
    """Get weather information for the specified city."""
    api_key = 'your_api_key_here'
    base_url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}"
    response = requests.get(base_url)
    data = response.json()
    if data['cod'] == 200:
        main = data['main']
        weather = data['weather'][0]
        temperature = main['temp']
        description = weather['description']
        speak(f"सध्या {city} मध्ये तापमान आहे {temperature - 273.15:.2f}°C आणि हवामान आहे {description}.")
    else:
        speak("क्षमस्व, मला हवामानाची माहिती मिळाली नाही.")
```

```python
def main():
    """Main function to run the assistant."""
    time.sleep(3)  # Pause before greeting
    wishme()

    while True:
        text = takecommand()
        
        if 'शुभ प्रभात' in text or 'शुभ दुपार' in text or 'शुभ संध्याकाळ' in text:
            speak('मी तुम्हाला कसे मदत करू शकतो?')
        
        elif 'विकिपीडिया मध्ये शोधा' in text or 'विकिपीडिया' in text:
            query = text.replace('विकिपीडिया', '').strip()
            speak('तुम्हाला मला सांगायचं आहे की वेबपेज उघडायचं आहे?')
            answer = takecommand()
            if 'कथन करा' in answer:
                results = wikipedia.summary(query, sentences=1, auto_suggest=False)
                speak('विकिपीडिया नुसार ' + results)
            elif 'वेबपेज' in answer:
                page = wikipedia.page(query, auto_suggest=False)
                webbrowser.open_new_tab(page.url)
        
        elif text == '':
            speak('क्षमस्व, तुम्ही काय म्हणाला ते पुन्हा सांगा.')
        
        elif 'गुगल मध्ये शोधा' in text:
            query = text.replace('गुगल मध्ये', '').replace('शोधा', '').strip()
            speak(query + ' गुगल मध्ये शोधत आहे.')
            webbrowser.open_new_tab('https://www.google.com/search?q=' + query)
        
        elif 'यूट्यूब मध्ये शोधा' in text:
            query = text.replace('यूट्यूब मध्ये', '').replace('शोधा', '').strip()
            speak(query + ' यूट्यूब मध्ये शोधत आहे.')
            webbrowser.open_new_tab('https://www.youtube.com/results?search_query=' + query)
        
        elif 'तुमचे नाव काय आहे' in text:
            speak('माझं नाव आहे Virtual Assistant.')
        
        elif 'हॅलो' in text:
            speak('हॅलो सर. मी तुम्हाला कसे मदत करू शकतो?')
        
        elif 'तुम्ही कसे आहात' in text:
            speak('मी ठीक आहे. तुम्हाला काही मदत हवी आहे का?')
        
        elif 'कोण तयार केले' in text:
            speak('माझे निर्माते Coder आहेत.')
        
        elif 'स्वतःची ओळख करून द्या' in text or 'तुम्ही कोण आहात' in text:
            speak('मी Virtual Assistant आहे. मी तुमचा वैयक्तिक संगणक सहाय्यक आहे.')
        
        elif 'धन्यवाद' in text:
            speak('स्वागत आहे! आणखी काही?')
        
        elif 'फेसबुक उघडा' in text:
            speak('फेसबुक उघडत आहे.')
            webbrowser.open_new_tab('https://www.facebook.com')
        
        elif 'इन्स्टाग्राम उघडा' in text:
            speak('इन्स्टाग्राम उघडत आहे.')
            webbrowser.open_new_tab('https://instagram.com')
        
        elif 'यूट्यूब उघडा' in text:
            speak('यूट्यूब उघडत आहे.')
            webbrowser.open_new_tab('https://www.youtube.com')
        
        elif 'गुगल क्रोम उघडा' in text:
            speak('गुगल क्रोम उघडत आहे.')
            os.startfile('https://www.google.com')
        
        elif 'स्पॉट उघडा' in text:
            speak('स्पॉटिफाय उघडत आहे.')
            os.system('spotify')
        
        elif 'संगीत वाजवा' in text:
            speak('ठीक आहे सर, तुमचं संगीत वाजवत आहे.')
            os.system('spotify')
            time.sleep(1)
            pyautogui.click(button='left')
            pyautogui.press('space')
            pyautogui.hotkey('alt', 'f4')
            while True:
                wakeup_txt = takecommand()
                if wakeup_txt == 'quit':
                    speak('बाय बाय सर. छान दिवस घालवा.')
                    sys.exit()
                elif 'pause' in wakeup_txt or 'play' in wakeup_txt:
                    os.system('spotify')
                    time.sleep(1)
                    pyautogui.press('space')
                    pyautogui.hotkey('alt', 'f4')
                elif 'स्पॉट बंद करा' in wakeup_txt:
                    os.system('TASKKILL /F /IM Spotify.exe')
        
        elif 'शटडाउन करा' in text:
            speak('ठीक आहे, संगणक बंद करत आहे.')
            os.system('shutdown /s /f')
            sys.exit()

        elif 'नोटपॅड उघडा' in text:
            speak('नोटपॅड उघडत आहे.')
            os.system('notepad')
        
        elif 'व्हि ऐस स्टुडिओ कोड उघडा' in text:
            speak('व्हिज्युअल स्टुडिओ कोड उघडत आहे.')
            os.system('code')
        
        elif 'वर्ड उघडा' in text:
            speak('मायक्रोसॉफ्ट वर्ड उघडत आहे.')
            os.system('start winword')
        
        elif 'एक्सेल उघडा' in text:
            speak('मायक्रोसॉफ्ट एक्सेल उघडत आहे.')
            os.system('start excel')
        
        elif 'पॉवर पॉइंट उघडा' in text:
            speak('मायक्रोसॉफ्ट पॉवरपॉइंट उघडत आहे.')
            os.system('start powerpnt')
        
        elif 'हवामान' in text:
            city = text.replace('हवामान', '').strip()
            get_weather(city)
        
        elif 'वेळ' in text:
            current_time = datetime.datetime.now().strftime("%H:%M:%S")
            speak(f"सध्या वेळ आहे {current_time}")
        
        elif 'शुभ रात्री' in text:
            speak('शुभ रात्री सर.')
        
        elif 'कॅमेरा उघडा' in text:
            speak('कॅमेरा उघडत आहे.')
            os.system('start microsoft.windows.camera:')
        
        elif 'शोध' in text:
            speak(text + ' गुगल, विकिपीडिया किंवा यूट्यूब मध्ये शोधा?')
            confirmation = takecommand()
            if 'गुगल' in confirmation:
                speak(text + ' गुगल मध्ये शोधत आहे.')
                webbrowser.open_new_tab('https://www.google.com/search?q=' + text)
            elif 'विकिपीडिया' in confirmation:
                speak('वेबपेज उघडायचं आहे का कथन करायचं आहे?')
                answer2 = takecommand()
                if 'कथन करा' in answer2:
                    results = wikipedia.summary(text, sentences=1, auto_suggest=False)
                    speak('विकिपीडिया नुसार ' + results)
                elif 'वेबपेज' in answer2:
                    page = wikipedia.page(text, auto_suggest=False)
                    webbrowser.open_new_tab(page.url)
            elif 'यूट्यूब' in confirmation:
                speak(text + ' यूट्यूब मध्ये शोधत आहे.')
                webbrowser.open_new_tab('https://www.youtube.com/results?search_query=' + text)
        
        else:
            speak('क्षमस्व, हे कार्य नोंदलेले नाही.')
```

```python
if __name__ == "__main__":
    main()
```
